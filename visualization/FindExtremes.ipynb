```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import argrelextrema
```

```python
x = np.linspace(0,19, 200)
y = np.sin(x) + 0.5 * np.cos(2 * x)


max_inxs = argrelextrema(y, np.greater)[0]
min_inxs = argrelextrema(y, np.less)[0]
```

```python

x_max = x[max_inxs] 
y_max = y[max_inxs] 
x_min = x[min_inxs] 
y_min = y[min_inxs] 


print("Local Maxima:")
for i, (xm, ym) in enumerate(zip(x_max, y_max)):
    print(f"Max {i+1}: x = {xm:.3f}, y = {ym:.3f}")

print("\nLocal Minima:")
for i, (xm, ym) in enumerate(zip(x_min, y_min)):
    print(f"Min {i+1}: x = {xm:.3f}, y = {ym:.3f}")

# Step 5: Plot the function and extrema
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Function')
plt.scatter(x_max, y_max, color='red', label='Maxima', zorder=5)
plt.scatter(x_min, y_min, color='blue', label='Minima', zorder=5)
plt.legend()
plt.grid(True)
plt.title(" Local Extrema")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
