# Heat Map

## Grid

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.linspace(-1,1, 100)
y = np.linspace(-1, 1, 100)
X, Y = np.meshgrid(x, y)
#print (X)
#print("--------")
#print(Y)
```

## function for heatmap




#### the function for heat map is
$Z=e−(x2+y2)cos⁡(4x)+tanh⁡(xy)+x3e−y2$

```python
Z= np.exp(-(X**2 + Y**2)) * np.cos(4 * X) + np.tanh(X * Y) + X**3 * np.exp(-Y**2)
```

```python
plt.figure(figsize=(12, 9))
sns.heatmap(Z, cmap='viridis', xticklabels=30, yticklabels=20, 
            cbar=True, cbar_kws={'label': 'Z Value', 'shrink': 0.6})
plt.title("Heatmap of Complex Function", fontsize=16, pad=20)
plt.xlabel("X", fontsize=12)
plt.ylabel("Y", fontsize=12)
#plt.xticks(np.arange(-1, 1, 40), [f"{val:.1f}" for val in x[::40]], rotation=45)
#plt.yticks(np.arange(-1, 1, 40), [f"{val:.1f}" for val in y[::40]])
plt.tight_layout()
plt.show()
```
