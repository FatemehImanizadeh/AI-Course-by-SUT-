```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import  make_classification
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
X, y = make_classification(n_samples=300, n_features=2, n_classes=2, n_redundant=0, n_informative=2, random_state=3)
plt.scatter(X[:,0], X[:,1], c = y, edgecolors= 'k' , s = 20)
plt.title("KNN with 2 Feature")
plt.xlabel("feature 1")
plt.ylabel("feature 2")
plt.grid(True)
plt.show()
x_train_val, x_test, y_train_val, y_test = train_test_split(X, y, test_size = 0.2, random_state = 3)
x_train, x_val, y_train, y_val = train_test_split(x_train_val, y_train_val, test_size=0.2, random_state= 3)
k_vals = np.arange(1,15,2)
best_k = None
best_val_score = 0
for k in k_vals:
    knn = KNeighborsClassifier(n_neighbors= k)
    knn.fit(x_train, y_train)
    y_val_predict = knn.predict(x_val)
    val_accuracy = accuracy_score(y_val, y_val_predict)
    print(f"k = {k}, validation accuracy = {val_accuracy:3f}")
    if  val_accuracy > best_val_score:
        best_k = k
        best_val_score = val_accuracy
print(f"best k = {best_k}, best validation accuracy = {best_val_score:3f}")
knn_final = KNeighborsClassifier(n_neighbors= best_k)
knn_final.fit(x_train_val, y_train_val)
y_test_predict = knn_final.predict(x_test)
test_accuracy = accuracy_score(y_test, y_test_predict)
print(f"test accuracy for best k({best_k}): {test_accuracy:3f}")
```
