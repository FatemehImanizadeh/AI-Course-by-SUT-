```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import  make_classification
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score
```

```python
X, y = make_classification(n_samples=300, n_features=2, n_classes=2, n_redundant=0, n_informative=2, random_state=344)
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 3)
k_vals = np.arange(1,15,1)
cross_val_scores = []
best_k = None
best_val_score = 0
for k in k_vals:
    knn = KNeighborsClassifier(n_neighbors= k)
    scores = cross_val_score(knn, x_train, y_train, cv = 5)
    cross_val_scores.append(np.mean(scores))
    print(f"k = {k}, cross validation accuracy = {np.mean(scores):3f}")
best_k = k_vals[np.argmax(cross_val_scores)]  
best_val_score = np.max(cross_val_scores)
knn_best = KNeighborsClassifier(n_neighbors= best_k)   
knn_best.fit(x_train, y_train)
y_test_predict = knn_best.predict(x_test)
print(f"best k = {best_k}, best validation accuracy = {best_val_score:3f}")
#preview real test data
plt.scatter(x_test[:,0], x_test[:,1], c = y_test, edgecolors= 'k' , s = 20)
plt.title("KNN Test Real")
plt.xlabel("feature 1")
plt.ylabel("feature 2")
plt.grid(True)
plt.show()
#preview predicted test data
plt.scatter(x_test[:,0], x_test[:,1], c = y_test_predict, edgecolors= 'k' , s = 20)
plt.title("KNN Test Predicted")
plt.xlabel("feature 1")
plt.ylabel("feature 2")
plt.grid(True)
plt.show()
```
