```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
```

# **linear regression**

```python
n = 1000
x = np.random.normal(3,5, size= n)
y = -25 * x  + 20 + np.random.normal(0,3,size = n)
```

```python
X = np.sum(x)
Y = np.sum(y)
XX = np.dot(x,x)
XY = np.dot(x,y)
b1 = (n*XY -X*Y)/(n*XX - X**2)
b0 = (Y - b1*X)/n
b1, b0 = round(b1, 2), round (b0, 2)
print(b1, b0)
```

**MSE LOSS**

```python
def MSE_Loss(b0, b1, x, y):
  pred = b0 + b1*x
  loss = np.sum((pred-y)**2/len(y))
  return loss
b0_vals = np.linspace(0, 40, 200)
b1_vals = np.linspace(-50, 0, 200)
Z = np.zeros((len(b0_vals), len(b1_vals)))
for i in range(len(b0_vals)):
  for j in range(len(b1_vals)):
   Z[i,j] =np.log(MSE_Loss(b0_vals[i], b1_vals[j], x, y))
```

**Plot**

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(b0_vals, b1_vals, Z)
B0, B1 = np.meshgrid(b0_vals, b1_vals)
ax.plot_surface(B0, B1, Z)
ax.set_xlabel('b0')
ax.set_ylabel('b1')
ax.set_zlabel('Loss')
plt.show()
```
