# **linear regression**

```python
n = 1000
x = np.random.normal(3,5, size= n)

y = -15 * x  + 20 + np.random.normal(0,3,size = n)
print (y)
```

```python
X = np.sum(x)
Y = np.sum(y)
XX = np.dot(x,x)
XY = np.dot(x,y)
b1 = (n*XY -X*Y)/(n*XX - X**2)
b0 = (Y - b1*X)/n
b1, b0 = round(b1, 2), round (b0, 2)
print(b1, b0)
```
